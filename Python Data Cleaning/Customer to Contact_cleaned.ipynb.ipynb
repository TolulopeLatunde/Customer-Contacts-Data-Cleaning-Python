```python
import pandas as pd
import re
```

```python
df = pd.read_excel(r"C:\Users\Admin\OneDrive\Desktop\drmaths\Python Projects\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
# Define function to clean phone numbers
def clean_phone_number(phone):
    if pd.isna(phone) or phone in ['N/a', 'NaN']:
        return None
    # Extract digits from phone number
    digits = re.sub(r'\D', '', phone)
    if len(digits) == 10:
        # Format as XXX-XXX-XXXX
        return f"{digits[:3]}-{digits[3:6]}-{digits[6:]}"
    return None  # If phone number doesn't have 10 digits

# Apply the cleaning function to the 'Phone_Number' column
df['Phone_Number'] = df['Phone_Number'].apply(clean_phone_number)

df
```

```python
# Define a function to split the address into Street, State, and Zip Code
def split_address(address):
    if pd.isna(address) or address == 'N/a':
        return pd.Series([None, None, None])
    
    # Split the address by the last comma
    parts = address.split(',')
    
    if len(parts) == 3:
        street_address = parts[0].strip()
        state = parts[1].strip()
        zip_code = parts[2].strip()
    elif len(parts) == 2:
        street_address = parts[0].strip()
        state = parts[1].strip()
        zip_code = None
    else:
        street_address = address.strip()
        state = None
        zip_code = None
    
    return pd.Series([street_address, state, zip_code])

# Apply the function to split the 'Address' column
df[['Street_address', 'State', 'Zip_code']] = df['Address'].apply(split_address)

# Drop the original 'Address' column if no longer needed
df = df.drop(columns=['Address'])

# Display the updated DataFrame
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df = df.replace('N/a','')
#df = df.replace('NaN','')


df=df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df

#Another way to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.to_excel("final_Customer call list.xlsx", index=False)
```
